```python
#Load the Pandas Liabrary. Must be installed on the system running this script
import pandas as pd
#Load the OS liabrary which comes with python installation
import os
#Load the csv liabrary which comes with every python installation
import csv
#Point the script to the directory containing the AS Interacton csv files downloaded directly from 5view
monthly_report = os.listdir("C:\\Users\\kingsley.idoko\\Videos\\Pandas\\New folder\\06_June")
#Loop through the CSV files in the directory already reference in the code above.
for file in monthly_report:
    #pick one CSV file at a time and point it to Pandas for processing
    formatted_file = f"New folder/06_June/{file}"
    #Pandas loads the the csv file for processing
    report = pd.read_csv(f'{formatted_file}') 
    #Picking the first Row and split it Into multiple columns. Because data from 5view are in series.this convert it to dataframe or tables
    output = report['AS Number 1;AS Number 2;Avg Total Throughput (bits/s);Avg 1 to 2 Throughput (bits/s);Avg 2 to 1 Throughput (bits/s);Max Total Throughput (bits/s);Max 1 to 2 Throughput (bits/s);Max 2 to 1 Throughput (bits/s);Connected Time (seconds)'].str.split(';', expand=True)
    # After spliting the row into multiple columns. You would need to specify the name of each columns
    output.columns = ['AS Number 1', 
                      'AS Number 2', 
                      'Avg Total Throughput (bits/s)', 
                      'Avg 1 to 2 Throughput (bits/s)', 
                      'Avg 2 to 1 Throughput (bits/s)', 
                      'Max Total Throughput (bits/s)', 
                      'Max 1 to 2 Throughput (bits/s)', 
                      'Max 2 to 1 Throughput (bits/s)', 
                      'Connected Time (seconds)']
    # Selecting only the column of interest.
    selected_columns = output[ ["AS Number 1",
                                "AS Number 2", 
                                "Max 1 to 2 Throughput (bits/s)",
                                "Max 2 to 1 Throughput (bits/s)",
                                "Max Total Throughput (bits/s)"]]
    #Each columns comes with data type of object. We need to convert it back to integers.
    selected_columns["AS Number 1"] = selected_columns["AS Number 1"].astype('int64')
    selected_columns["AS Number 2"] = selected_columns["AS Number 2"].astype('int64')
    selected_columns["Max 1 to 2 Throughput (bits/s)"] = selected_columns["Max 1 to 2 Throughput (bits/s)"].astype('int64')
    selected_columns["Max 2 to 1 Throughput (bits/s)"] = selected_columns["Max 2 to 1 Throughput (bits/s)"].astype('int64')
    selected_columns["Max Total Throughput (bits/s)"] = selected_columns["Max Total Throughput (bits/s)"].astype('int64')
    top_twelve = selected_columns.sort_values("Max Total Throughput (bits/s)", ascending=False).head(15)
    #Open and Load the BGP AS number database. In this case it was in the same directory with where the script is been ran
    db = open("asn.txt", 'r')
    #Load the ASN database into a varriable.
    db1 = csv.reader(db)
    # loop through the ASN Database and match where the ASN matches and swap the name to ASN name.
    for row in db1:        
        var1 = int(row[0]) 
        #If the CSV is for google CACHE then ASN OF 0 does not mean MainOne But GOOGLE CACHE. 
        if file == 'GOOGLE CACHE.csv' and var1 == 0:
            var2 = 'GOOGLE CACHE'         
        #If the CSV is for NETFLIX CACHE then ASN OF 0 does not mean MainOne But NETFLIX CACHE. 
        elif file == 'NETFLIX_CACHE.csv' and var1 == 0:
            var2 = 'NETFLIX CACHE'
        # if the above conditions does not match. simple set ASN 0 a MainOne
        else:
            var2 = f'{row[1]}({var1})'
        # Iterate through each row of AS Number 1 then AS Number 2 and resolve the AS Number with the AS Name.
        for index, rows in top_twelve.iterrows():
            if top_twelve.loc[index, "AS Number 1"] == var1:
                top_twelve.loc[index, "AS Number 1"] = var2
            if top_twelve.loc[index, "AS Number 2"] == var1:
                top_twelve.loc[index, "AS Number 2"] = var2 
    # Load the slice data into a new varriale
    final_report = top_twelve
    # Save the new CSV file to specified directory using the same Import name. Optionally take out the index values.
    final_report.to_csv(f"Finish_Product/{file}", index=False)
    
```
